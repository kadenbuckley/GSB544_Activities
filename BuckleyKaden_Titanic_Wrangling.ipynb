```python
import numpy as np
import pandas as pd
from plotnine import *
# Import the titanic dataset here
data_dir = "https://dlsun.github.io/pods/data/"
df_titanic = pd.read_csv(data_dir + "titanic.csv")
#load titanic data
df_titanic
```

## 1. Filter the data to include passengers only. Calculate the joint distribution (cross-tab) between a passenger's class and where they embarked.

```python
# Code here
#Filter data to include passengers only
df_po = df_titanic[(df_titanic["class"] == "1st") | (df_titanic["class"] == "2nd") | (df_titanic["class"] == "3rd")]

#Create Cross-tab for class and embarked
pd.crosstab(df_po["class"], df_po["embarked"])
```

## 2. Using the joint distribution that calculated above, calculate the following:

* the conditional distribution of their class given where they embarked
* the conditional distribution of where they embarked given their class

Use the conditional distributions that you calculate to answer the following quesitons:

* What proportion of 3rd class passengers embarked at Southampton?
* What proportion of Southampton passengers were in 3rd class?

```python
# Code here
#Calculate cond. dist. of class given where they embarked
cge = pd.crosstab(df_po["class"], df_po["embarked"], normalize=True)
print(cge)

#Calculate cond. dist. of where they embarked given class
egc = pd.crosstab(df_po["embarked"], df_po["class"], normalize=True)
print(egc)
```

## 3. Make a visualization showing the distribution of a passenger's class, given where they embarked.

Discuss the pros and cons of using this visualization versus the distributions you calculated before, to answer the previous questions.

```python
# Code here
#Passenger column
df_titanic["passenger_class"] = df_titanic["class"].map({
    "1st": "passenger",
    "2nd": "passenger",
    "3rd": "passenger",
    "victualling crew": "crew",
    "engineering crew": "crew",
    "deck crew": "crew"
})

df_passenger = df_titanic[df_titanic["passenger_class"]=="passenger"]


plot1 = (ggplot(df_passenger, aes("embarked",fill = "class"))
+ geom_bar(position='fill')
+ labs(title='Passenger Class | Embark Location', x='Embark Location', y='Proportion'))
plot1
```

Providing a visualization in this case gives a much better sense of the whole picture. Calculating only numbers, as we did before, makes things much harder to interpret. With a plot such as this one, we can now compare proportions of passengers based on their class, given all of the embark locations. A con to a visualiation such as this one is a loss of exact information, whereas the calculations provided previously show exact proprtions.
